### **Import Libraries**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Car_Evaluation_Clear.csv")
```

```python
sns.set_style(style="whitegrid")
```

```python
sns.countplot(data=df,x="class",palette="Set2")
plt.show()
```

There is a clear imbalance between classes. Therefore, it is advisable to treat the data.

### **Balanced Dataset**

```python
def load_inputs_outputs():
    
    return df.drop(["class"],axis="columns"),df["class"]

X,y=load_inputs_outputs()
```

```python
from imblearn.over_sampling import SMOTE
```

Smote will help us create new examples. So that the classes are perfectly balanced.

```python
smote=SMOTE()
```

```python
X_balanced,y_balanced=smote.fit_resample(X,y)
```

```python
df_balanced=pd.concat([X_balanced,y_balanced],axis="columns")
```

```python
df_balanced["class"].value_counts()
```

```python
fig,(ax) = plt.subplots(1,1,figsize=(20,8))

ax.set_title("Balanced Dataset")
sns.countplot(data=df_balanced,x="class",palette="Set2",ax=ax)
plt.show()
```

Both classes are perfectly balanced.

#### **Save Dataset**

```python
df_balanced.to_csv("Car_Evaluation_Balanced.csv",index=False)
```
